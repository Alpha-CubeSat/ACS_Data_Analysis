```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("test.csv")
time = data.iloc[:,0]
#maybe not xyz
w_x = data.iloc[:,1]
w_y = data.iloc[:,2]
w_z = data.iloc[:,3]
```

```python
plt.figure(dpi=150)
plt.plot(time, w_x, label = "w_x")
plt.plot(time, w_y, label = "w_y")
plt.plot(time, w_z, label = "w_z")
plt.legend()
plt.show()
```
