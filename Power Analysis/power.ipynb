```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array(pd.read_csv("data3.txt"))

t = np.arange(0,data.shape[0]*0.2*100-1,0.2*100)

abs_current1 = np.abs(data[:,[0]])
pt_err1 = np.abs(data[:,[1]])

abs_current2 = np.abs(data[:,[0+2]])
pt_err2 = np.abs(data[:,[1+2]])

abs_current3 = np.abs(data[:,[0+4]])
pt_err3 = np.abs(data[:,[1++4]])
```

```python
plt.figure(dpi=150)
current_avg1  = np.mean(abs_current1)
current_avg_line1 = np.ones(abs_current1.shape)*current_avg1

current_avg2  = np.mean(abs_current2)
current_avg_line2 = np.ones(abs_current2.shape)*current_avg2

current_avg3  = np.mean(abs_current3)
current_avg_line3 = np.ones(abs_current3.shape)*current_avg3
#25%
plt.plot(t,abs_current1)
plt.plot(t,abs_current2)
plt.plot(t,abs_current3)

plt.plot(t,current_avg_line1,'r',label="I_avg(100%): "+str(current_avg1)+" mA")
plt.plot(t,current_avg_line2,'g',label="I_avg(15%): "+str(current_avg2)+" mA")
plt.plot(t,current_avg_line3,'b',label="I_avg(25%): "+str(current_avg3)+" mA")
plt.xlabel("Time (s)")
plt.ylabel("Current (mA)")

plt.title("ACS Simulation pointing current(magnitude) |I_z| output")

plt.legend()
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(t,pt_err1*100, label="error (100%)")
plt.plot(t,pt_err2*100, label ="error (15%)")
plt.plot(t,pt_err3*100, label ="error (25%)")

plt.xlabel("Time (s)")
plt.ylabel("Pointing Error (%)")

plt.title("ACS Simulation pointing error")

plt.legend()
plt.show()
```

```python
Vcc=4.0
print("Assume Vcc = "+str(Vcc)+"V, Average Power Consumption(100%): "+str('{0:.3f}'.format(0.001*current_avg1*Vcc))+" W")
print("Assume Vcc = "+str(Vcc)+"V, Average Power Consumption(15%): "+str('{0:.3f}'.format(0.001*current_avg2*Vcc))+" W")
print("Assume Vcc = "+str(Vcc)+"V, Average Power Consumption(25%): "+str('{0:.3f}'.format(0.001*current_avg3*Vcc))+" W")
print("Power Consumption Ratio: "+str('{0:.3f}'.format((current_avg2/current_avg1)*100))+" %")
```
