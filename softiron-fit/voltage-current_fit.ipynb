```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
data = pd.read_csv("flight/Flight-Voltagechange.txt")

v = data.iloc[:,1]
current = data.iloc[:,2]

start = 0
end = len(data)
```

```python
plt.figure(dpi=150)
plt.plot(v[start:end], current[start:end], '.',label = "current")
plt.xlabel("V")
plt.ylabel("A")
plt.title("Current vs Voltage")
plt.legend()
plt.show()
```

```python

coeffs = np.polyfit(v[start:end], current[start:end]/0.194, 1)


# Create polynomial functions from the coefficients
poly= np.poly1d(coeffs)

# Generate a range of PWM values for plotting the fit
fit = np.linspace(3, 4, 500)

# Plot the original data
plt.figure(dpi=150)
plt.plot(v[start:end], current[start:end]/0.194, '.', label="Data")


# Plot the fitted curves
plt.plot(fit, poly(fit), label="Fit")
plt.xlabel("V")
plt.ylabel("A")
plt.title("Linear Fit of V vs.A")


plt.legend()
plt.show()

print("coeff "+ str(coeffs))
```
