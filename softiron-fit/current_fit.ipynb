```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
data = pd.read_csv("flight/Flight-PWMchange.txt")

pwm =  data.iloc[:,0]
current =  data.iloc[:,2]

start = 0
end = len(data)
```

```python
plt.figure(dpi=150)
plt.plot(pwm[start:end], current[start:end], '.',label = "current")

plt.xlabel("PWM")
plt.ylabel("A")
plt.title("Current")
plt.legend()
plt.show()
```

```python

coeffs = np.polyfit(current[start:end], pwm[start:end], 2)


# Create polynomial functions from the coefficients
poly = np.poly1d(coeffs)

# Generate a range of PWM values for plotting the fit
current_fit = np.linspace(0, 0.2, 500)

# Plot the original data
plt.figure(dpi=150)
plt.plot(current[start:end], pwm[start:end], '.', label="Data")


# Plot the fitted curves
plt.plot(current_fit, poly(current_fit), label="Fit")

plt.ylabel("pwm")
plt.xlabel("Current (A)")
plt.title("2nd Order Polynomial Fit of PWM vs. Current")


plt.legend()
plt.show()

print("PWM to Current: "+ str(coeffs))
```
