```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("LP_raw.txt")
mag_x_raw = df.iloc[:,0]
mag_x_smooth = df.iloc[:,3]

mag_y_raw = df.iloc[:,0+1]
mag_y_smooth = df.iloc[:,3+1]

mag_z_raw = df.iloc[:,0+2]
mag_z_smooth = df.iloc[:,3+2]
z_current = df.iloc[:,12]

w_x_raw = df.iloc[:,6]
w_y_raw = df.iloc[:,7]
w_z_raw = df.iloc[:,8]

w_x_smooth = df.iloc[:,6+3]
w_y_smooth = df.iloc[:,7+3]
w_z_smooth = df.iloc[:,8+3]


mag_z_v = mag_z_smooth/np.sqrt(mag_x_smooth*mag_x_smooth+mag_y_smooth*mag_y_smooth+mag_z_smooth*mag_z_smooth)
x = np.arange(0,mag_x_raw.size)
t = x*0.25

df
```

```python
plt.plot(t[range_a:range_b], z_current[range_a:range_b],label="pointing(z_mag) current(mA)")
plt.xlabel("sec")
plt.ylabel("mA")
plt.legend()
```
