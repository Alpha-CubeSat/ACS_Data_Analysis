```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
data = pd.read_csv("ekfsensor.txt")

time =  data.iloc[:,0]

mag_x= data.iloc[:,1]
mag_y = data.iloc[:,2]
mag_z = data.iloc[:,3]
wx = data.iloc[:,4]
wy = data.iloc[:,5]
wz = data.iloc[:,6]

mag_xf= data.iloc[:,7]
mag_yf = data.iloc[:,8]
mag_zf = data.iloc[:,9]
wxf = data.iloc[:,10]
wyf = data.iloc[:,11]
wzf = data.iloc[:,12]


start =0
end = 100 #len(data)
```

```python
plt.figure(dpi=150)
plt.plot(time[start:end], wxf[start:end], label = "wxf")
plt.plot(time[start:end], wxf[start:end], label = "wxf")

plt.plot(time[start:end], wyf[start:end], label = "wyf")
plt.plot(time[start:end], wyf[start:end], label = "wyf")

plt.plot(time[start:end], wzf[start:end], label = "wzf")
plt.plot(time[start:end], wzf[start:end], label = "wzf")

plt.xlabel("sec")
plt.ylabel("rad/s")
plt.title("angular velocity")
plt.legend()
plt.show()
```
