```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
data = pd.read_csv("test.txt")

time =  data.iloc[:,0]

wx = data.iloc[:,1]
wy = data.iloc[:,2]
wz = data.iloc[:,3]

cx  = data.iloc[:,5]
cy  = data.iloc[:,6]
cz  = data.iloc[:,7]
error = data.iloc[:,4]

start = 0
end = len(data)
```

```python
time = time[start:end]

plt.figure(dpi=150)
plt.plot(time, wx[start:end], label = "wx")
plt.plot(time, wy[start:end], label = "wy")
plt.plot(time, wz[start:end], label = "wz")



plt.xlabel("sec")
plt.ylabel("rad/s")
plt.title("angular velocity")
plt.legend()
```

```python
plt.figure(dpi=150)
plt.plot(time[start:end],cx[start:end],alpha=0.7, label = "current x")
plt.plot(time[start:end],cy[start:end],alpha=0.7 ,label = "current y")
plt.plot(time[start:end],cz[start:end],alpha=0.7 ,label = "current z")



plt.xlabel("sec")
plt.ylabel("A")
plt.title("current")
plt.legend()
plt.show()
```
