```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
data = pd.read_csv("test.txt")

time =  data.iloc[:,0]

wx = data.iloc[:,1]
wy = data.iloc[:,2]
wz = data.iloc[:,3]

error = data.iloc[:,4]

start = 0
end = len(data)
```

```python


time = time[start:end]

plt.figure(dpi=150)
plt.plot(time, wx[start:end], label = "wx")
plt.plot(time, wy[start:end], label = "wy")
plt.plot(time, wz[start:end], label = "wz")



plt.xlabel("sec")
plt.ylabel("rad/s")
plt.title("angular velocity")
plt.legend()
plt.show()

```

```python

time = time[start:end]

plt.figure(dpi=150)
plt.plot(time, error[start:end], label = "error")

plt.xlabel("sec")
plt.ylabel("deg")
plt.title("pointing error")
plt.legend()
plt.show()
```
